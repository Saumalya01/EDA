```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/drive/MyDrive/Data Analysis/data.csv')
data
```

Data Pre-Processing

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.diagnosis.unique()
```

```python
data.drop(columns=['Unnamed: 32'],inplace=True)
```

```python
data.head()
```

```python
#Checking the distribution of diagnosis variable
data['diagnosis'].value_counts()
```

```python
#Converting the diagnosis variable into 1 & 0's (where 1 --> Benign & 0 --> Malignant)
data['diagnosis'].replace(['B','M'],[1,0],inplace=True)
```

```python
#Finding the mean value of every column values for both the labels in diagnosis
data.groupby('diagnosis').mean()
```

```python
data.drop(columns=['id'],inplace=True)
```

Seperating the features and target

```python
data
```

```python
X=data.drop(columns='diagnosis',axis=1) #axis=1 means going horizontally
Y=data['diagnosis']
```

```python
Y
```

Splitting the data into training and testing data

```python
#All the corresponding labels in diagnosis column for X_test is stored in Y_train, while same for X_test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)    #test_size=0.2 means 20% of the data is reserved for testing
```

Model Training 

```python
model=LogisticRegression()
```

```python
#Training the Logistic Regression model using Training data(i.e, X_train , Y_train)
model.fit(X_train,Y_train)
```

Model Evaluation using Accuracy Score

```python
X_train_prediction=model.predict(X_train)  #Predicting the Diagnosis label (i.e, 1 or 0)from the X_train values
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)   #To check how much accurate our prediction model is
```

```python
print("The Accuracy on the training data is =", training_data_accuracy)
```

```python
#Accuracy on Test data
X_test_prediction=model.predict(X_test) 
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print("The Accuracy on the testing data is =", test_data_accuracy)
```

Building a Prediction System
