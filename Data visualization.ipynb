# Data Visualization using matplotlib,seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

**%matplotlib inline** command is used to ensure our graphs remain intact in our notebook

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples);
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
#Putting axis
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
#Plotting multiple lines
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896 ]
plt.plot(years,apples,marker='o') #markers
plt.plot(years,oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title('Crops yield in Himachal')
plt.legend(['Apples','Oranges']); #Adding legend
```

**More on matplotlib ---> https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html**

**We can use fmt argument for a shorthand for marker style,colour,shape and all and the meaning of 'o--r' is marker type is round,marker line is dotted and marker color is red**

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title('Crops yield in Himachal')
plt.legend(['Apples','Oranges']);
```

**Improving default styles with seaborn**

```python
sns.set_style("darkgrid")
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title('Crops yield in Himachal')
plt.legend(['Apples','Oranges']);
```

**Scatter Plot**

```python
import pandas as pd
flowers_df=pd.read_csv('iris.csv')
flowers_df
```

```python
plt.plot(flowers_df.SepalLengthCm,flowers_df.SepalWidthCm);
```

```python
#Adding colours to different species and performing a scatterplot
plt.title("Sepal Dimensions")
sns.scatterplot(x=flowers_df.SepalLengthCm,
                y=flowers_df.SepalWidthCm,
                hue=flowers_df.Species,
                s=50);
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.SepalWidthCm);
```

```python
# Specifying the boundaries of each bin/interval
import numpy as np
plt.hist(flowers_df.SepalWidthCm,bins=np.arange(2,5,0.25));
```

```python
#Bins with unequal size
plt.hist(flowers_df.SepalWidthCm,bins=[1,3,4,4.5]);
```

```python
#Drwaing seperate histograms for different species of flower
setosa_df = flowers_df[flowers_df.Species == 'Iris-setosa']
virginica_df = flowers_df[flowers_df.Species == 'Iris-virginica']
versicolor_df = flowers_df[flowers_df.Species == 'Iris-versicolor']
```

```python
plt.hist(setosa_df.SepalWidthCm, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(virginica_df.SepalWidthCm, alpha=0.4, bins=np.arange(2, 5, 0.25));
#plt.hist(versicolor_df.SepalWidthCm, alpha=0.4, bins=np.arange(2, 5, 0.25));

plt.title("Comapring Distribution of Sepal Width")
plt.legend(['Setosa','Virginica']);
```

```python
#Stacking multiple histogram on top of other
plt.title('Distribution of Sepal Width')

plt.xlabel('Sepal Width')
plt.ylabel('Frequency')

plt.hist([setosa_df.SepalWidthCm, versicolor_df.SepalWidthCm, virginica_df.SepalWidthCm], 
         bins=np.arange(2, 5, 0.25),
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

**Plotting multiple charts in a grid**

```python
fig,axes=plt.subplots(2,3,figsize=(12,9));
plt.tight_layout(pad=2)

# Using the axes for plotting in the 1st graph
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r');
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')

#Passing the axes into Seaborn in the 2nd graph
axes[0,0].plot(years,oranges,'o--r'); #It's like a matrix in the grid and can be used as like plt
sns.scatterplot(x=flowers_df.SepalWidthCm,
                y=flowers_df.SepalLengthCm,
                hue=flowers_df.Species,
                s=100,
                ax=axes[0,1]);

#Using axes to plot histogram for the 3rd graph
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.SepalWidthCm, versicolor_df.SepalWidthCm, virginica_df.SepalWidthCm], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img);
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])
```
